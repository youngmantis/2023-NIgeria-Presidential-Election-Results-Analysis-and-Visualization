```python
import pandas as pd

#load the xlsx file 
df = pd.read_excel('/content/drive/MyDrive/2023_Nigeria_Presidential_Election_Results.xlsx')

df
```

```python
#compute the total valid votes of all the parties
df['TOTAL_VALID_VOTES'] = df.loc[:, ['APC', 'PDP', 'LP', 'NNPP', 'OTHERS']].sum(axis = 1)

#compute the vote percentage of parties by state
df['APC_VOTES_PERCENT'] = (100/df['TOTAL_VALID_VOTES']) * df['APC']
df['PDP_VOTES_PERCENT'] = (100/df['TOTAL_VALID_VOTES']) * df['PDP']
df['LP_VOTES_PERCENT'] = (100/df['TOTAL_VALID_VOTES']) * df['LP']
df['NNPP_VOTES_PERCENT'] = (100/df['TOTAL_VALID_VOTES']) * df['NNPP']
df['OTHERS_VOTES_PERCENT'] = (100/df['TOTAL_VALID_VOTES']) * df['OTHERS']


df
```

```python
#states won by 25% and above by parties
APC_25 = df['APC_VOTES_PERCENT'] >= 25
PDP_25 = df['PDP_VOTES_PERCENT'] >= 25
LP_25 = df['LP_VOTES_PERCENT'] >= 25
NNPP_25 = df['NNPP_VOTES_PERCENT'] >= 25
OTHERS_25 = df['OTHERS_VOTES_PERCENT'] >= 25

print('APC number of states won by 25% and above:', APC_25.value_counts(), '\n'
      'PDP number of states won by 25% and above:', PDP_25.value_counts(), '\n'
      'LP number of states won by 25% and above:', LP_25.value_counts(), '\n'
      'NNPP number of states won by 25% and above:', NNPP_25.value_counts(), '\n'
      'Other parties number of states won by 25% and above:', OTHERS_25.value_counts(), '\n')
```

```python
#to calculate the mov, create a new dataframe (df4) for the party columns
df4 = df[['APC', 'PDP', 'LP', 'NNPP', 'OTHERS']]

df4.head()
```

```python
#get the second highest votes for rows  in df4
df4['Second_Highest_Votes'] = df4.apply(lambda row: row.nlargest(2).values[-1], axis = 1)

df4['Second_Highest_Votes'].head()
```

```python
#get the highest votes for rows in df4
df4['Highest_Votes'] = df4.max(axis=1)
df4['Highest_Votes'].head()
```

```python
#view the dataframe
df4.head()
```

```python
#add the total valid votes column to df4
df4['TOTAL_VALID_VOTES'] = df['TOTAL_VALID_VOTES']

df4.head()
```

```python
#compute the margin of victory
df4['MOV_Percent'] = (((df4['Highest_Votes'] - df4['Second_Highest_Votes'])/df4['TOTAL_VALID_VOTES']) * 100)
df4['STATE'] = df['STATE']

df4
```

```python
#close states i.e. states with narrow mov
df4['MOV_Under_One_Percent'] = df4['MOV_Percent'] < 1
df4['MOV_Between_1_and_5_Percent'] = df4['MOV_Percent'].between(1,5)
df4['MOV_Between_5_and_10_Percent'] = df4['MOV_Percent'].between(5,10)

df4
```

```python
#states with wide mov
df4['MOV_10_Percent_and_above'] = df4['MOV_Percent'] >= 10
df4['MOV_20_Percent_and_above'] = df4['MOV_Percent'] >= 20
df4['MOV_30_Percent_and_above'] = df4['MOV_Percent'] >= 30
df4['MOV_40_Percent_and_above'] = df4['MOV_Percent'] >= 40
df4['MOV_50_Percent_and_above'] = df4['MOV_Percent'] >= 50
df4['MOV_60_Percent_and_above'] = df4['MOV_Percent'] >= 60
df4['MOV_70_Percent_and_above'] = df4['MOV_Percent'] >= 70
df4['MOV_80_Percent_and_above'] = df4['MOV_Percent'] >= 80
df4['MOV_90_Percent_and_above'] = df4['MOV_Percent'] >= 90

df4
```
